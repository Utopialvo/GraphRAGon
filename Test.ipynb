```python
import sys
sys.path.append('./src')

import logging
from graph_rag import GraphRAG, GraphRAGConfig

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# Загрузка конфигурации из YAML
config = GraphRAGConfig.from_yaml("config.yaml")

graph_rag = GraphRAG(config)
graph_rag.clear()
```

```python
large_text = """
Глава 1. Утром Иван нарубил дрова в сарае. Затем Варвара растопила печь.
Пока печь разгоралась, Иван пошёл на речку и стал ловить рыбу.
В это же время Варвара замесила тесто для хлеба.
После того как Иван вернулся с рыбалки, он почистил рыбу.
Варвара испекла хлеб, когда печь уже хорошо прогрелась.
Затем они вместе накрыли на стол и поужинали.
Перед ужином Иван вымыл руки.
Вечером, после ужина, Варвара убрала со стола, а Иван читал книгу.
На следующий день Иван снова рубил дрова, а Варвара готовила завтрак.
Иван отнёс дрова в сарай. Варвара испекла пироги.
"""

print("=== Загрузка текста и построение графа ===")
passage_ids = graph_rag.process_text(large_text)
print(f"Создано {len(passage_ids)} чанков")
print(f"Метрики: {graph_rag.get_metrics()}\n")
```

```python
questions = [
    "Что делал Иван?",
    "Что делала Варвара?",
    "Какие действия выполняли Иван и Варвара вместе?",
]

print("\n=== Диалог с системой (GoT включён, параллельно) ===")
for q in questions:
    ans = graph_rag.ask(q, use_got=True)
    print(f"\nВопрос: {q}")
    print(f"Ответ: {ans}")
```

```python
print("\n=== Диалог с системой (GoT выключён, параллельно) ===")
for q in questions:
    ans = graph_rag.ask(q, use_got=False)
    print(f"\nВопрос: {q}")
    print(f"Ответ: {ans}")
```

```python
print("\n--- История диалога ---")
for entry in graph_rag.get_history():
    print(f"Вопрос: {entry['question']}\nОтвет: {entry['answer']}\n")
```
